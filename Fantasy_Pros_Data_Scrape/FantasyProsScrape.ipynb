```python
## Pull data from the table on https://www.fantasypros.com/nfl/stats/qb.php?year=2022&range=week&week=1
## Concatenate PASSING RUSHING MISC with the respective fields below ex: "RANK", "PLAYER", "PASSING_CMP", "PASSING_ATT", "PASSING_PCT", "PASSING_YDS", "PASSING_Y/A", 
##                               "PASSING_TD", "PASSING_INT", "PASSING_SACKS", "RUSHING_ATT", "RUSHING_YDS", "RUSHING_TD", 
##                               "MISC_FL", "MISC_FPTS", "MISC_FPTS/G", "MISC_FUM", "MISC_ROST".
## The first two rows are not needed so we can remove them.
# df = pd.DataFrame(data[2:], columns=['RANK', 'PLAYER', 'PASSING_CMP', 'PASSING_ATT', 'PASSING_PCT', 'PASSING_YDS', 'PASSING_Y/A', 
#                                  'PASSING_TD', 'PASSING_INT', 'PASSING_SACKS', 'RUSHING_ATT', 'RUSHING_YDS', 'RUSHING_TD',
#                                     'MISC_FL', 'MISC_FPTS', 'MISC_FPTS/G', 'MISC_FUM', 'MISC_ROST'])
## Create a function called qb_data using the code above that takes in year and week in as paramaters and returns the dataframe for that year.  
## Also insert a column for year in the 2nd column of the dataframe.

def get_QB_data(year, week):
    url = 'https://www.fantasypros.com/nfl/stats/qb.php?year=' +str(year) + '&range=week&week='+ str(week)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    table = soup.find('table', {'id': 'data'})
    table_rows = table.find_all('tr')
    data = []
    for tr in table_rows:
        td = tr.find_all('td')
        row = [tr.text for tr in td]
        data.append(row)
    df = pd.DataFrame(data[2:], columns=['RANK', 'PLAYER', 'PASSING_CMP', 'PASSING_ATT', 'PASSING_PCT', 'PASSING_YDS', 'PASSING_Y/A', 
                                 'PASSING_TD', 'PASSING_INT', 'PASSING_SACKS', 'RUSHING_ATT', 'RUSHING_YDS', 'RUSHING_TD',
                                    'MISC_FL', 'MISC_FPTS', 'MISC_FPTS/G', 'MISC_FUM', 'MISC_ROST'])
    df.insert(1, 'YEAR', year)
    df.insert(2, 'WEEK', week)
    df['PLAYER_ID'] = df['PLAYER'].str.extract('(\d+)')
    return df

import pyodbc
conn = pyodbc.connect('Driver={SQL Server};'
                        'Server=localhost\SQLEXPRESS;'
                        'Database=FantasyFootball;'
                        'Trusted_Connection=yes;')
cursor = conn.cursor()
sql = '''
CREATE TABLE NFL.QB_STATS(
    RANK INT,
    YEAR INT,
    WEEK INT,
    PLAYER VARCHAR(255),
    PLAYER_ID INT,
    PASSING_CMP INT,
    PASSING_ATT INT,
    PASSING_PCT FLOAT,
    PASSING_YDS INT,
    PASSING_YD_Avg FLOAT,
    PASSING_TD INT,
    PASSING_INT INT,
    PASSING_SACKS INT,
    RUSHING_ATT INT,
    RUSHING_YDS INT,
    RUSHING_TD INT,
    MISC_FL INT,
    MISC_FPTS FLOAT,
    MISC_FPTS_G FLOAT,
    MISC_FUM INT,
    MISC_ROST INT
)
'''
cursor.execute(sql)
cursor.commit()
# cursor.execute('CREATE SCHEMA NFL;')
# cursor.commit()
# cursor.execute('CREATE SCHEMA Sleeper;')
# cursor.commit()
cursor.close()


```
